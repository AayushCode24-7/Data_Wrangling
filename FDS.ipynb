Step 2: Data Wrangling

```python
df_cleaned = df.copy()
```

```python
# Cleaning The Data

df_cleaned = df.drop(columns=['Unnamed: 0']).copy()
df_cleaned
```

```python
# Removing The Duplicates From The Track ID

df_cleaned = df.drop_duplicates(subset=['track_id'])
df_cleaned
```

```python
# Dropped The Null Row

df_cleaned = df_cleaned.dropna()
df_cleaned
```

```python
# Explicit Content True/False

df_cleaned['explicit'] = df_cleaned['explicit'].astype(bool)
df_cleaned
```

```python
df_cleaned['duration_ms'] = df_cleaned['duration_ms']/60000
df_cleaned
```

```python
print(f"Data type of df_cleaned: {type(df_cleaned)}")
df_cleaned
```

```python
print(f"Columns available: {df_cleaned.columns.tolist()}")
df_cleaned
```

Step 3: Visualization (Graphs)

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Scatter Plot---> every dot is a song.
# if dots are increasing diagonally then it has corelation

sample = df_cleaned.sample(1000)

plt.figure(figsize=(10,6))
plt.scatter(sample['energy'], sample['danceability'], alpha=0.3, s=10)
plt.xlabel('Energy')
plt.ylabel('Danceability')
plt.title('Energy vs Danceability')
plt.show()
```

```python
# Bar Chart - Top 10 Genres by Average Popularity

top_genres = df_cleaned.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top_genres.index, top_genres.values, color='skyblue', edgecolor='black')
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.title('Top 10 Genres by Average Popularity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Line Chart - Average Energy and Danceability across Top 15 Genres

top15 = df_cleaned.groupby('track_genre')[['energy','danceability']].mean().sort_values(by='energy', ascending=False).head(15)

plt.figure(figsize=(14,6))
plt.plot(top15.index, top15['energy'], marker='o', label='Energy', color='red', linewidth=2)
plt.plot(top15.index, top15['danceability'], marker='s', label='Danceability', color='blue', linewidth=2)
plt.xlabel('Genre')
plt.ylabel('Average Score')
plt.title('Energy vs Danceability across Top 15 Genres')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# HISTOGRAM--->
# it is a right-skewed graph.

plt.figure(figsize=(10,6))
plt.hist(df_cleaned['popularity'], bins=30, color='purple', edgecolor='black', alpha=0.7)
plt.xlabel('Popularity')
plt.ylabel('Number of Tracks')
plt.title('Distribution of Track Popularity')
plt.tight_layout()
plt.show()
```

```python
# Box Plot - Popularity Distribution across Top 10 Genres

top10_genres = df_cleaned['track_genre'].value_counts().head(10).index
top10 = df_cleaned[df_cleaned['track_genre'].isin(top10_genres)]

plt.figure(figsize=(14,6))
sns.boxplot(x='track_genre', y='popularity', data=top10, palette='Set2')
plt.xlabel('Genre')
plt.ylabel('Popularity')
plt.title('Popularity Distribution across Top 10 Genres')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
