Step 2: Data Wrangling

```python
df_cleaned = df.copy()
```

```python
# Cleaning The Data

df_cleaned = df.drop(columns=['Unnamed: 0']).copy()
df_cleaned
```

```python
# Removing The Duplicates From The Track ID

df_cleaned = df.drop_duplicates(subset=['track_id'])
df_cleaned
```

```python
# Dropped The Null Row

df_cleaned = df_cleaned.dropna()
df_cleaned
```

```python
# Explicit Content True/False

df_cleaned['explicit'] = df_cleaned['explicit'].astype(bool)
df_cleaned
```

```python
df_cleaned['duration_ms'] = df_cleaned['duration_ms']/60000
df_cleaned
```

```python
print(f"Data type of df_cleaned: {type(df_cleaned)}")
df_cleaned
```

Step 3: Visualization (Graphs)

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Scatter Plot---> every dot is a song.
# if dots are increasing diagonally then it has corelation
# Seaborn library.

top5 = df_cleaned[df_cleaned['track_genre'].isin(['pop','rock','hip-hop','jazz','classical'])]
sample = top5.sample(1000)
plt.figure(figsize=(10,6))
sns.scatterplot(x='energy', y='danceability', hue='track_genre', data=sample, alpha=0.5, s=30)
plt.title('Energy vs Danceability (Top 5 Genres)')
plt.legend(title='Genre')
plt.tight_layout()
plt.show()
```

```python
# Bar Chart - Top 10 Genres by Average Popularity
# matplotlib libeary.

top_genres = df_cleaned.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
plt.bar(top_genres.index, top_genres.values, color='skyblue', edgecolor='black')
plt.xlabel('Genre')
plt.ylabel('Popularity')
plt.title('Top 10 Genres Popularity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Line Chart - Average Energy and Danceability across Top 15 Genres.
# Matpllotlib

top15 = df_cleaned.groupby('track_genre')[['energy','danceability']].mean().sort_values(by='energy', ascending=False).head(15)
plt.figure(figsize=(14,6))
plt.plot(top15.index, top15['energy'], marker='o', label='Energy', color='red', linewidth=2)
plt.plot(top15.index, top15['danceability'], marker='s', label='Danceability', color='blue', linewidth=2)
plt.xlabel('Genre')
plt.ylabel('Score')
plt.title('Energy vs Danceability (Top 15 Genres)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# HISTOGRAM--->
# it is a right-skewed graph.
# seaborn

plt.figure(figsize=(10,6))
sns.histplot(df_cleaned['popularity'], bins=30, kde=True, color='pink')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.title('Distribution of Track Popularity')
plt.tight_layout()
plt.show()
```

```python
# Box Plot - Popularity Distribution across Top 10 Genres
# Seaborn

top5_genres = df_cleaned['track_genre'].value_counts().head(5).index
top5 = df_cleaned[df_cleaned['track_genre'].isin(top5_genres)].sample(500)
plt.figure(figsize=(10,6))
sns.boxplot(x='track_genre', y='popularity', data=top5, palette='Set2')
plt.xlabel('Genre')
plt.ylabel('Popularity')
plt.title('Popularity Distribution across Top 5 Genres')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
